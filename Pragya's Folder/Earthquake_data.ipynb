```python
import pandas as pd
import numpy as np
```

```python
earthquake_df = pd.read_csv("clean_data1.csv")
earthquake_df.head()
```

```python
earthquake_df1 = earthquake_df.loc[earthquake_df['YEAR'] >= 1999]

earthquake_df1.head()
```

```python
# Create new table
earthquake_df2 = earthquake_df1[["YEAR", "MONTH", "DAY","HOUR", "MINUTE", "SECOND",
                           "EQ_PRIMARY", "COUNTRY","LOCATION_NAME","STATE","LATITUDE", "LONGITUDE", "DEATHS", "Depth"]]
earthquake_df2.head()


earthquake_df2.reset_index(drop=True)
```

```python
earthquake_df2.to_csv("earthquake_data.csv", index=False)
```
