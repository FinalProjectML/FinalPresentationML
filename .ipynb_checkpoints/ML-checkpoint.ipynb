```python
# I am predicting the eq_primary which is the magnitude of earthquakes using linear regression
```

```python
le = preprocessing.LabelEncoder()
region = le.fit_transform(list(file2["REGION_CODE"]))
file2['region'] = region
#file2.head()
```

```python
file2['REGION_CODE'].unique()
```

```python
dummy = pd.get_dummies(file2['REGION_CODE'])
```

```python
dummy.head()
```

```python
file2 = pd.concat([file2, dummy], axis=1)
```

```python
file2.head()
```

```python
#del file2['region']
file2.head()
```

```python
#file2.corr()
corrmat = file2.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(22,15))
g = sns.heatmap(file2[top_corr_features].corr() > .70, annot=True, cmap="RdYlGn")
```

```python
(file2.corr())> 0.70
```

```python
predict = 'REGION_CODE'
```

```python
x = np.array(file2.drop([predict, 'DEATHS', 'INJURIES', 'YEAR', 'MONTH'],1))
```

```python
y = np.array(file2[predict])
x_train, x_test , y_train, y_test = sklearn.model_selection.train_test_split(x,y, test_size=0.25)
```

```python
model = KNeighborsClassifier(n_neighbors=11)
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```
